**前提**

* Zを確率変数（random variable）とする.
* Zがとりうる値のそれぞれに確率を与える関数がzの確率分布関数（probability distribution function）である.

**離散の場合**

* もしZが離散の場合、確率分布は確率質量関数（probability mass function）と呼ばれる.

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style("whitegrid")

import scipy.stats as stats
```

```python
poi = stats.poisson
lambda_ = [1.5, 4.25]
colors = ["#348ABD", "#A60628"]
```

```python
a = np.arange(16) # 0, 1, 2, 3, 4, 5 ... 15
```

```python
plt.figure(figsize=(12.5, 6))

plt.bar(a, poi.pmf(a, lambda_[0]), color=colors[0], label="$\lambda = %.1f$" % lambda_[0], alpha=0.6, edgecolor=colors[0], lw="3")
plt.bar(a, poi.pmf(a, lambda_[1]), color=colors[1], label="$\lambda = %.1f$" % lambda_[1], alpha=0.6, edgecolor=colors[1], lw="3")

plt.xticks(a + 0.4, a)
plt.legend()
plt.ylabel("Probability of $k$")
plt.xlabel("$k$")
plt.title("Probability mass function of a Poisson random variable")
plt.show()
```
